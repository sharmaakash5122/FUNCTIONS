6.Explain the concept of generators in Python and how they are defined.

- Generators are functions that yield one item at a time instead of returning all values at once.

Defined using the yield keyword.

Example:

def count_up_to(n):

    i = 1
    while i <= n:
        yield i
        i += 1

for num in count_up_to(3):

    print(num)

10.What is the difference between `map()`, `reduce()`, and `filter()` functions in Python?

- map()

- 1. Applies a given function to all items in an iterable (like a list).

- 2. Returns a map object (which can be converted to a list).

Example: map(lambda x: x * 2, [1, 2, 3]) results in [2, 4, 6].

- filter()

 - 1. Filters items in an iterable based on a function that returns True or False.

- 2. Returns a filter object (which can also be converted to a list).

Example: filter(lambda x: x &gt; 1, [1, 2, 3]) results in [2, 3].

- reduce()

- 1. Applies a rolling computation to sequential pairs of values in an iterable.
- 2. Requires importing from the functools module.
- 3. Returns a single cumulative value.

Example: reduce(lambda x, y: x + y, [1, 2, 3]) results in 6.

```python
'''1.Write a Python function that takes a list of numbers as input and returns the sum of all even numbers in
the list.'''

def sum_even_numbers(numbers):
    even_sum = 0
    for num in numbers:
        if num % 2 == 0:
            even_sum += num
    return even_sum

# Example:
nums = [10, 15, 20, 25, 30]
print("Sum of even numbers:", sum_even_numbers(nums))
```

```python
'''4. Write a Python function that checks if a given number is prime or not from 1 to 200.'''

def is_prime(num):
    if num < 2:
        return False
    for i in range(2, int(num**0.5) + 1):
        if num % i == 0:
            return False
    return True

for i in range(1, 201):
    if is_prime(i):
        print(i, end=" ")
```

```python
'''5.Create an iterator class in Python that generates the Fibonacci sequence up to a specified number of
terms.'''
class Fibonacci:
    def __init__(self, n):
        self.n = n
        self.a, self.b = 0, 1
        self.count = 0

    def __iter__(self):
        return self

    def __next__(self):
        if self.count < self.n:
            value = self.a
            self.a, self.b = self.b, self.a + self.b
            self.count += 1
            return value
        else:
            raise StopIteration

# Example:
for num in Fibonacci(10):
    print(num, end=" ")

```

```python
'''6.Write a generator function in Python that yields the powers of 2 up to a given exponent.'''

def power_of_two(limit):
    for i in range(limit + 1):
        yield 2 ** i

for val in power_of_two(5):
    print(val, end=" ")
```

```python
'''7.Implement a generator function that reads a file line by line and yields each line as a string.'''

def read_file(filename):
    try:
        with open(filename, 'r') as file:
            for line in file:
                yield line.strip()   # yields each line one by one
    except FileNotFoundError:
        print("Error: File not found!")

# Example use:
for line in read_file("example.txt"):
    print(line)


```

```python
'''8.Use a lambda function in Python to sort a list of tuples based on the second element of each tuple.'''

data = [(1, 5), (3, 1), (2, 8), (4, 2)]
sorted_data = sorted(data, key=lambda x: x[1])
print(sorted_data)
```

```python
'''9. Write a Python program that uses `map()` to convert a list of temperatures from Celsius to Fahrenheit.'''

celsius = [0, 10, 20, 30, 40]
fahrenheit = list(map(lambda c: (9/5)*c + 32, celsius))
print(fahrenheit)
```

```python
'''10. Create a Python program that uses `filter()` to remove all the vowels from a given string.'''

def remove_vowels(text):
    vowels = "aeiouAEIOU"
    return ''.join(filter(lambda ch: ch not in vowels, text))

print(remove_vowels("Beautiful Python"))
```

```python
'''11. Imagine an accounting routine used in a book shop. It works on a list with sublists, which look like this:
Write a Python program, which returns a list with 2-tuples. Each tuple consists of the order number and the
product of the price per item and the quantity. The product should be increased by 10,- â‚¬ if the value of the
order is smaller than 100,00 â‚¬.

Write a Python program using lambda and map.'''

# Sample order data â†’ [OrderNo, ItemName, Quantity, PricePerItem]
orders = [
    [34587, "Learning Python", 4, 40.95],
    [98762, "Programming in C", 5, 56.80],
    [77226, "Data Structures", 3, 32.95],
    [88112, "Algorithms", 3, 24.99]
]

# Use lambda + map
# For each order â†’ (OrderNo, adjusted_total)
invoice_totals = list(map(
    lambda order: (
        order[0],
        round(order[2] * order[3] if order[2] * order[3] >= 100 else order[2] * order[3] + 10, 2)
    ),
    orders
))

print(invoice_totals)
```
